Our initial step will be to address the genre list.



```python
genre = imdb[['Genre', 'IMDB_Rating']]
genre
```

We will split all the genre and collectively calculate the imdb provided of each genre respectively . Then we will replace the mixed genres from the values in the table by calculating their mean .

```python
genre['Genre'] = genre['Genre'].str.split(', ')
genre = genre.explode('Genre')

genre
```

```python
Genre_final=genre.groupby('Genre')['IMDB_Rating'].mean().reset_index()
Genre_final.set_index('Genre', inplace=True)

# Define get_imdb_rating function
def get_imdb_rating(genre):
    if genre in Genre_final.index:
        return Genre_final.loc[genre, 'IMDB_Rating']
    else:
        return 0  # Return 0 if genre is not found

# Define calculate_average_rating function
def calculate_average_rating(genres):
    if isinstance(genres, str):
        genres_list = genres.split(', ')
        total_sum = 0
        count = 0
        
        for genre in genres_list:
            total_sum += get_imdb_rating(genre)  # Use the genre variable
            count += 1
        
        if count > 0:
            return total_sum / count
        else:
            return 0  # Return 0 if no valid genres are found
```

```python
# Apply the function to the 'Genre' column
imdb['Genre'] = imdb['Genre'].apply(calculate_average_rating)

imdb
```

The genre column has now been successfully transformed into numerical data.

```python
Year_Gross=imdb[['Released_Year','Gross']]
```

```python
Year_Gross
```

```python
Year_Gross_cleaned=Year_Gross.dropna()

Year_Gross_cleaned
```

```python
Year_Gross_cleaned=Year_Gross_cleaned.groupby('Released_Year')['Gross'].mean().reset_index()

Year_Gross_cleaned
```

```python
Year_Gross_cleaned.set_index('Released_Year',inplace=True)
```

```python
def get_gross(year):
    if year in Year_Gross_cleaned.index:
        return Year_Gross_cleaned.loc[year, 'Gross']
    else:
        return 0

def get_gross_nan(imdb):
    def fill_gross(row):
        if pd.isna(row['Gross']):
            return get_gross(row['Released_Year'])
        else:
            return row['Gross']
    
    imdb['Gross'] = imdb.apply(fill_gross, axis=1)
    return imdb
```

```python
imdb = get_gross_nan(imdb)
imdb
```

We will now proceed to apply a scikit-learn model to evaluate the accuracy of our dataframe.

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
```

```python
input_imdb_data=imdb[['Released_Year','Genre','Meta_score','Director','Star1','Star2','Star3','Star4','No_of_Votes','Gross']]
output_imdb=imdb['IMDB_Rating']
```

```python
x_train,x_test,y_train,y_test=train_test_split(input_imdb_data,output_imdb,test_size=0.2)
```

```python
Alg1 = DecisionTreeRegressor(max_depth=5)
Alg1.fit(x_train, y_train)

print(Alg1.score(x_test,y_test))

Alg2 = RandomForestRegressor(n_estimators=100, max_depth=5)
Alg2.fit(x_train, y_train)
print(Alg2.score(x_test,y_test))
```
