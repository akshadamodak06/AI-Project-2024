Since there is no variation in the output due to 'No' we drop it 

```python
data.drop('No',axis=1,inplace=True)
```

```python
sb.heatmap(data,cmap='BrBG')
```

Also since the std deviation of 'X1 transaction date', 'X5 latitude' and 'X6 longitude' is less than zero hence we drop them too.

```python
data.drop(['X1 transaction date','X5 latitude', 'X6 longitude'],axis=1,inplace=True)
```

```python
sb.heatmap(data,cmap='BrBG')
```

We now train the model and test it.

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
train=LinearRegression()
train_x,test_x,train_y,test_y=train_test_split(data.drop('Y house price of unit area',axis=1),data['Y house price of unit area'],test_size=0.01)
train.fit(train_x,train_y) 
print(train.score(test_x,test_y))
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y=train_test_split(data.drop('Y house price of unit area',axis=1),data['Y house price of unit area'],test_size=0.01)
clf1=RandomForestRegressor()
clf1.fit(train_x,train_y)
print(clf1.score(test_x,test_y))
```
